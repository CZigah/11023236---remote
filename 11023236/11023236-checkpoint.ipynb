## Locating missing values in data

```python
data.isnull().sum()
```

```python
mid_fa = data_stat.fixed_acidity.loc['50%']

mid_va = data_stat.volatile_acidity.loc['50%']

mid_c = data_stat.chlorides.loc['50%']

mid_fsd = data_stat.free_sulfur_dioxide.loc['50%']

mid_d = data_stat.density.loc['50%']

mid_q = data_stat.quality.loc['50%']
```

```python
fData = data.copy()
```

```python
fData.fixed_acidity = data.fixed_acidity.fillna(mid_fa)

fData.volatile_acidity = data.volatile_acidity.fillna(mid_va)

fData.chlorides = data.chlorides.fillna(mid_c)

fData.free_sulfur_dioxide = data.free_sulfur_dioxide.fillna(mid_fsd)

fData.density = data.density.fillna(mid_d)

fData.quality = data.quality.fillna(mid_q)
```

```python
fData.isnull().sum()
```

## Searching for duplicate values

```python
fData.duplicated()
```

```python
fData=fData.drop_duplicates()
```

```python
fData.duplicated()
```

## Histogram; using pH

```python
bins= np.arange(0.990,1.003,0.001)

mat.figure(figsize=(7,5))
mat.hist(fData.density, bins=13, rwidth=0.6) 
mat.xticks()

mat.title('Various densities of red wine',fontdict={'fontstyle': "italic"})
mat.xlabel('Density')
mat.ylabel('red wine')

mat.show()
```

#### From the histogram above, the majority of red wines have densities that range from 0.996 to 0.997

#### Likewise, few red wines have densities ranging from 0.990-0.991

## Scatter Diagram; Volatile Acidity against pH

```python
%matplotlib inline
```

```python
mat.figure(figsize=(8,6))
mat.title('volatile acidity over pH',fontdict={'fontstyle': "italic"})
mat.scatter(x="pH", y="volatile_acidity",data = fData, c="g", alpha= 0.6) 
mat.xlabel('pH')
mat.ylabel('Volatile acidity')

mat.legend()
```

#### The scatter plot shows the relationship between pH and volatile acidity. The x-axis represents pH, ranging from 2.8 to 4.0, while the y-axis represents volatile acidity, ranging from 0 to 1.6. The majority of data points are clustered between a pH of approximately 3.0 to 3.8 and a volatile acidity of around 0.4 to 1.0, indicating a concentration in this range. There is no clear trend or pattern visible, as the data points are densely packed without forming any distinct shape or linearity, suggesting a complex relationship between these two variables that isn't easily summarized by a simple correlation or trend line. The color coding indicates all points represent volatile_acidity data, shown in green dots scattered across the plot area.
